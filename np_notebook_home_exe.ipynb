## <u>Exercises 1 :</u>

- creates a random matrix in range 10 - 99,<br>
  <u>then calculates the sum of all rows & columns .</u>

```python
random_matrix = np.random.randint(10, 99, (4,3))
np_matrix_sum = 0
print(random_matrix)
for row in random_matrix:
    for rand_array_num in row:
        np_matrix_sum += rand_array_num
print(np_matrix_sum)
```

## <u>Exercises 2 :</u>
- showing how to sum all the even elemnts in numpy array
  <br> <u>using np.sum() .</u>


```python
random_array_len_10 = np.random.randint(0, 11, (10))
print(random_array_len_10)
even_num_count = np.sum(random_array_len_10 % 2 == 0)
print(even_num_count)
```

## <u>Exercises 3 :</u>
- showing how to arange, reshap, replace elemnts, <br><u>sum elements and their avarage in a numpy matrix .</u>

```python
#part 1.
large_np_array = np.arange(0, 100)
print(large_np_array)

#part 2.
matrix_10x10 = large_np_array.reshape(10, 10)
print(matrix_10x10)

#part 3.
matrix_10x10[matrix_10x10 > 50] = -1
print(matrix_10x10)

#part 4.
matrix_10x10_sum = np.sum(matrix_10x10)
print(matrix_10x10_sum)

#part 5.
matrix_10x10_mean = np.mean(matrix_10x10)
print(matrix_10x10_mean)
    
```

## <u>Exercises 4 :</u>
- showing how to use linspace
  <br><u>and how to take the arrays elements to make a new array with them using conditions.</u>

```python
#part 1.
lin_array = np.linspace(0, 100, 10)
print(lin_array)

#part 2.
new_array = lin_array[(lin_array > 50) | (lin_array < 20)]
print(new_array)

#part 3.
reshaped_array = new_array.reshape(7, 1)
print(reshaped_array)
```

## <u>Exercises 5 :</u>
- showing how to make np zeros matrix
  <br><u>and how to replace elements in the matrix with randomly generated numbers .</u>

- <u>i used random librarry because its more simple to generate 1 random number and choose it compared to np.random.randint() .</u>
  

```python
#part 1.
np_4x4_zeros = np.zeros((4,4), dtype=int)
print(np_4x4_zeros)

#part 2.
np_4x4_zeros[0][0] = random.randint(10, 100)

np_4x4_zeros[1][3] = random.randint(10, 100)

np_4x4_zeros[-1][-1] = random.randint(10, 100)
print(np_4x4_zeros)
```
